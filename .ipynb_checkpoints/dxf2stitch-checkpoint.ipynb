```python
MyEntities=[ele for ele in dxf.blocks['block 2']]#Found a bunch of splines
```

```python
#can I find out more about these splines. They should be circles.
#for entity in MyEntities:
 #   if entity.is_closed:
  #      print 'Closed'
   # else:
    #    print 'Open'
```

```python
stitchxdata=[]
stitchydata=[]
for entity in MyEntities:
    #print entity.controlpoints  #How can the x,y center be extracted from these splines?
    #print '\n'
    xval=0
    yval=0
    for point in entity.controlpoints: #what if I just find the centroid of each spline--good enough?
        xval=xval+point[0]
        yval=yval+point[1]
    xval=xval/len(entity.controlpoints)*25.4 #change it to mm
    yval=yval/len(entity.controlpoints)*25.4
    #print xval*25.4,yval*25.4
    #print '\n'
    stitchxdata.append(xval)
    stitchydata.append(yval)
        
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#getting this from here http://nbviewer.ipython.org/github/ipython/ipython/blob/1.x/examples/notebooks/Part%203%20-%20Plotting%20with%20Matplotlib.ipynb
plt.figure(1,figsize=(6,6))
plt.plot(stitchxdata,stitchydata,'b.')
plt.hold(True)
plt.plot(stitchxdata,stitchydata,'r-')
```
