```python
crawzacc=[]#arrays to hold z accceleration data
fordzacc=[]
with open ("3_26_2018SampleData.txt", "r") as myfile:
    data = myfile.readlines()
for line in data:
    #print item
    if 'Craw' in line:
        crawzacc.append(line[26:31])#this is easier if you separate data with a delimiter character,
        #you may be able to open right in excel if you use a comma or tab 
    if 'Ford' in line:
        fordzacc.append(line[26:31])
#check out data[270:290] when the x accel has 2 digits (rarely), everything is shifted over by one character
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
crawzacc[120:140] #There are some spikes in the CCraw data, check up on a spike, what is goin on
```

```python
#Convert from strings to floats for plotting
crawzaccfloat=[float(c) for c in crawzacc[1:]]#there is something it does not like at 0

#I'm just lop off first digit of any value greater than 99.9 bc it is coming from data occasionally
#getting shoved over by one character.
#To really fix, add a delimiter to the code that's collecting data 
import numpy as np
crawzaccWithoutSpikes=[c-100.*(np.fix(c/100.)) for c in crawzaccfloat]

fordzaccfloat=[float(f) for f in fordzacc[1:]]
```

```python
plt.figure(figsize=(10,4))
plt.subplot(2,1,1)
plt.plot(crawzaccWithoutSpikes[1:500])
plt.title('CCraw - Skipping')
plt.ylabel('Z accel reading m/s^2')
#plt.hold(True)
plt.subplot(2,1,2)
plt.plot(fordzaccfloat[1:500],'r')
plt.title('DFord - Jogging')
plt.xlabel('Row number')
plt.ylabel('Z accel reading m/s^2')
```
