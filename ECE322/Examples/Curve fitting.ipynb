```python
np.random.seed(12)#what is this - check the docs
x=np.linspace(0,1,20)
y=np.cos(x) +0.3*np.random.rand(20)
p=np.polyfit(x,y,1)  #What is the 3rd argument, check the docs
#The last term sets the polynomial order. 1st order = line, 2nd order 
#is Ax^2 +Bx +C  etc
t=np.linspace(0,1,200)
yfit=np.polyval(p,t)
plt.plot(x,y,'o',t,yfit,'-')
```

```python
#What if we change around 3 to something else
#Was it really a quadratic function?

#How about if we kinda know the underlying function is a sine or cosine? 
#Ther is a curve fitting package
```

```python
np.random.seed(0) #change this around to see different random data
x_data = np.linspace(-5,5)
y_data=2.9 * np.sin(1.5*x_data)+np.random.normal(size=50)

plt.scatter(x_data,y_data)
```

```python
from scipy import optimize #sounds awesome
def test_func(x,a,b):
    return a*np.sin(b*x)
params, params_covariance=optimize.curve_fit(test_func,x_data,y_data,p0=[2,2])
print(params)
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_data,y_data,label='Data')
plt.plot(x_data,test_func(x_data,params[0],params[1]),label='Fitted function')
plt.legend(loc='best')
plt.show()
```
