```python
#Parking
RedYearlyCost=590
BlueYearlyCost=258
GreenYearlyCost=121
MeterDailyCost=1.75*8  #Parking meter (must move every 4 hours)
VisitorDaypass =4
TicketCost=20
WorkdaysPerYear=261
import numpy as np
from matplotlib import pyplot as plt

T=np.linspace(1,WorkdaysPerYear,WorkdaysPerYear)
RedCostPerDay=[RedYearlyCost/t for t in T]
BlueCostPerDay=[BlueYearlyCost/t for t in T]
GreenCostPerDay=[GreenYearlyCost/t for t in T]
TicketCostPlotline=TicketCost*np.ones((WorkdaysPerYear,1))
OneTenthTicketCostPlotline=TicketCostPlotline/10
MeterCostPlotline=MeterDailyCost*np.ones((WorkdaysPerYear,1))
VisitorCostPlotline=VisitorDaypass*np.ones((WorkdaysPerYear,1))
%matplotlib inline
plt.figure(figsize=(8,5))

plt.hold('true')
plt.plot(T,RedCostPerDay,'r-')
plt.plot(T,BlueCostPerDay,'b-')
plt.plot(T,GreenCostPerDay, 'g-')
plt.plot(T,MeterCostPlotline,'m-')
plt.plot(T,TicketCostPlotline,'k--')
plt.plot(T,OneTenthTicketCostPlotline,'k--')
plt.plot(T,VisitorCostPlotline,'y-')
plt.title('Cost of various university parking plans')
plt.xlabel('number of workdays parked')
plt.ylabel('cost per day ($)')
plt.ylim(0,30)
plt.xlim(0,150)
plt.savefig('parking.eps')
```
