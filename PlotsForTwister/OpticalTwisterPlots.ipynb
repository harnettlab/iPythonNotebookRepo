```python
#Optical fiber twister k=2, n=24, 0.8 mm diameter fiber coated in Elastosil M4641, then put 
#on top of the kevlar-embedded membrane before the top was coverd with Eco Flex 10
import numpy as np
```

```python
#Data collected as I ramped the pressure up to 1.5 psi, held 5 s, down to "0" hold 5 sec, 5 times
#Each data point is 200 ms apart.
IntensityData= np.loadtxt('1pt5PSI_5xRepeat.csv', delimiter=',',skiprows=1)
LightSensorVolts=IntensityData[:,3]
Times=np.arange(0.2,len(LightSensorVolts)*0.2,0.2)#Timebase in seconds
len(LightSensorVolts),len(Times)
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot(Times,LightSensorVolts)
plt.xlabel('Time (seconds)')
plt.ylabel('Light Sensor Signal, Volts')#typically 10^6 V/A, this is from the TSL-12T integrated sensor
plt.title('Five cycles from 0 up to 1.5 psi, hanging out 5 seconds at each end')
#As the fiber stretches with inflation, the signal drops.
plt.savefig('5PressureCyclesOpticalSignal.eps')
```
