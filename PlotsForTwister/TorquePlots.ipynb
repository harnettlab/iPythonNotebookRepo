```python
#Plots of torque vs angle from data collected by ckh lab book 3/30/17
import numpy as np
#most of this deals with the k3n18 membrane bc it had the largest angle range
```

```python
#Data collected as I clamped the membrane holder at an angle and inflated to a pressure.
#0 degrees is horizontal, right on the scale; it should exert the most torque---
#unless high forces start to rip the Loctite from the silicone or constrain it somehow
#I suspect somehting like htat started to happen on the k=3 sample, so on the k=1 sample
#I stayed below 10 g because of this possibility. 10g felt like I was gonna tear it off
#The "np.nan" values mean I didn't measure it bc of potential ripping (>10g), 
#for example at small angles and large pressures,
#or the stick never reached the scale
#for example with large angles and low pressures.

#you can cut and paste these arrays into Matlab but the comments need % instead of #
k3n18_sticklength=0.14 # stick length in meters
k3n18_angle=[0,18,37.5,65,85,105,125,130]#starting points in degrees
k3n18_p175_weights=[5.1,6.1,6.65,7.94,7.2,5.7,2.61,1.7]#1.75 psi, weights in grams
k3n18_p150_weights=[4.83,5.9,6.62,5.51,6.0,3.67,1.39,0.66]#1.5 psi, weights in grams
k3n18_p125_weights=[4.42,3.52,5.43,3.45,4.7,1.69,np.nan,np.nan]#1.25 psi, weights in grams -- uncert is about 0.8 g
k3n18_p100_weights=[2.75,2.49,3.2,1.76,1.7,np.nan,np.nan,np.nan]#1 psi, weights in grams
k3n18_p075_weights=[1.61,1.16,1.17,np.nan,np.nan,np.nan,np.nan,np.nan]#0.75 psi, weights in grams


#The k1n32 membrane didn't have a big region where both the pressure and angle were within range
#It might have been interesting to analyze it at a few closely spaced angles around 45-60 degrees
#where it was both in the "safe" torque region <10g * sticklength
#and able to hit the scale. a rotation stage would make those expts go a lot faster
k1n32_sticklength=0.13 #stick length in meters
k1n32_angle=[0,20,45,65,90]
k1n32_p175_weights=[np.nan,np.nan,np.nan,4.13,np.nan]
k1n32_p150_weights=[np.nan,np.nan,np.nan,2.9,np.nan]
k1n32_p125_weights=[np.nan,np.nan,9.43,1.03,np.nan]
k1n32_p100_weights=[np.nan,np.nan,5.9,np.nan,np.nan]
k1n32_p075_weights=[np.nan,8.66,1.72,np.nan,np.nan]
k1n32_p050_weights=[8.08,2.6,np.nan,np.nan,np.nan]

#and in my lab book, where is k2 n24?..I only have spring constant measurements for it. 
#That will go in the spring constant notebook
```

```python
#convert weights to torques in N m, edit out "np.array" () to translate to matlab
#0.009807 Newtons per gram
k3n18_p075_torques=np.array(k3n18_p075_weights)*0.009807*k3n18_sticklength
k3n18_p100_torques=np.array(k3n18_p100_weights)*0.009807*k3n18_sticklength
k3n18_p125_torques=np.array(k3n18_p125_weights)*0.009807*k3n18_sticklength
k3n18_p150_torques=np.array(k3n18_p150_weights)*0.009807*k3n18_sticklength
k3n18_p175_torques=np.array(k3n18_p175_weights)*0.009807*k3n18_sticklength
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
plt.plot(k3n18_angle,k3n18_p075_torques,'y.')
plt.hold('true')
plt.plot(k3n18_angle,k3n18_p100_torques,'m.')
plt.plot(k3n18_angle,k3n18_p125_torques,'g.')
plt.plot(k3n18_angle,k3n18_p150_torques,'b.')
plt.plot(k3n18_angle,k3n18_p175_torques,'r.')

#draw some lines to see what is going on better
plt.plot(k3n18_angle,k3n18_p075_torques,'y--')
plt.plot(k3n18_angle,k3n18_p100_torques,'m--')
plt.plot(k3n18_angle,k3n18_p125_torques,'g--')
plt.plot(k3n18_angle,k3n18_p150_torques,'b--')
plt.plot(k3n18_angle,k3n18_p175_torques,'r--')

plt.xlabel('Start angle (degrees)')
plt.ylabel('Torque exerted(Nm)')
plt.title('Torque vs angle, k3n18 membrane y:0.75 psi, m: 1psi g: 1.25 psi b: 1.5 psi r: 1.75 psi')
#plt.savefig('TorqueVsAngle_k3n18.eps')
```
